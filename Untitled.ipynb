```python
import re
source = open('Themerchantofvenice.txt')
```

```python
def stringsearch(merchant):
    if type(merchant)!= str: 
        raise TypeError("Not a string")
    result = re.findall("\w*at\\b", merchant)
    atends = []
    for element in result: 
        if len(element) > 3:
            atends.append(element)
    return atends
```

```python
venice = source.read()
```

```python
stringsearch(venice)
```
